```python
#Page 36

fig, ax = plt.subplots(figsize=(11, 5))
#spines
ax.spines[['top', 'right']].set_visible(False)
ax.spines[['left','bottom']].set_color('grey')
ax.tick_params(color ='grey' )
#axis limit
ax.set_ylim(0,500)
ax.set_yticklabels(['$'+str(int(x)) for x in ax.get_yticks()],size=12,color = 'grey')
ax.set_xlim(2007.5,2014)
ax.set_xticks([x for x in range(2008,2015)],color ='grey')
ax.set_xticklabels([x for x in range(2008,2015)],color ='grey')
#highlighted line
ax.plot(df.T, linewidth=2.2,color = 'grey')
ax.plot(df.iloc[:2,2:].T,linewidth = 2.2,color ='darkblue')
#marker
ax.lines[2].set_markevery([2])
ax.lines[2].set_marker('o')
ax.lines[2].set_markersize(9)
ax.lines[2].set_markerfacecolor('darkblue')
#add legend
for i in range(df.shape[0]):
    line = ax.lines[i]
    for j in line.get_xydata():
        if str(j[1]) != 'nan':
            ax.text(j[0]-0.15,j[1]-10,df.index[i][-1],color='grey',size=12,alpha=0.8,weight='bold')
            break
#add label
ax.set_title('Average Retail Product Price per Year',size=18,color = 'grey')
```

```python
#1 tạo dataframe
women = {'Hong Kong': 87.6,
        'Japan': 87.5,
        'Switzerland': 85.5,
        'Singapore': 85.7,
        'Italy': 85.4,
        'Spain': 86.1,
        'Australia': 85.3,
        'Iceland': 84.4,
        'Israel': 84.4,
        'South Korea': 85.8
        }
men = { 'Hong Kong': 81.8,
        'Japan': 81.3,
        'Switzerland': 81.7,
        'Singapore': 81.4,
        'Italy': 81.1,
        'Spain': 80.7,
        'Australia': 81.3,
        'Iceland': 81.3,
        'Israel': 81.1,
        'South Korea': 79.7
        }
```

```python
import pandas as pd
import numpy as np
```

```python
li={'Women':pd.Series(women),'Men':pd.Series(men)}
life_exp_top10 = pd.DataFrame(li)
life_exp_top10
```

```python
#2 tạo cột trung bình
life_exp_top10['Average']=(life_exp_top10['Women']+life_exp_top10['Men'])/2
life_exp_top10
```

```python
#3 tạo cột rank
rank = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9, 9],index= men.keys())
life_exp_top10['Rank']=rank
life_exp_top10
```

```python
#4 thêm dòng 
# new_row={'Women':84.4,'Men':80.9,'Average':(84.4+80.9)/2}
# life_exp_top10 = life_exp_top10.append(pd.DataFrame([new_row],index=['Sweden'],columns=life_exp_top10.columns))
# life_exp_top10
```

```python
#4 theêm dòng
life_exp_top10.loc['Sweden']=[84.4,80.9,(84.4+80.9)/2,None]
life_exp_top10
```

```python
#update rank
life_exp_top10["Rank"] = life_exp_top10["Average"].rank(method='dense',ascending=False)
life_exp_top10
```

```python
#5 xóa dòng
life_exp_top10=life_exp_top10.drop('Italy')
life_exp_top10
```

```python
#6 xóa cột
life_exp_top10=life_exp_top10.drop(columns='Average') # thêm bỏ [] ksao
# life_exp_top10=life_exp_top10.drop('Average',axis=1)
life_exp_top10
```

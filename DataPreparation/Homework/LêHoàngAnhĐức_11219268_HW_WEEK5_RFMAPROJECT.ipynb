```python
#drop duplicated row
df=df.drop_duplicates()
df
```

```python
#delete records with the value of Quantity <=0
df=df[df['Quantity']>0]
df
```

```python
#drop records with null CustomerID
df=df[df['CustomerID'].notnull()]
df
```

```python
#Change type of InvoiceDate from object to date
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df['CustomerID']=df['CustomerID'].astype('int')
df.info()
```

```python
#Tạo df customer_df với columns sẵn 'CustomerID'
customer_df=pd.DataFrame(list(map(int,df['CustomerID'].sort_values().unique())))
customer_df
```

```python
#Create columns TotalRevenue
df['TotalRevenue']=df['Quantity']*df['UnitPrice']
df
```

```python
#Create frequency_df
frequency_df=df.groupby(['CustomerID'])['InvoiceNo'].agg([('Frequency','nunique')])
frequency_df
```

```python
#Create monetary_df
monetary_df=df.groupby(['CustomerID'])['TotalRevenue'].agg([('Monetary','sum')])
monetary_df
```

```python
#Create recency_df
recency_df=df['InvoiceDate'].max()-df.groupby('CustomerID')['InvoiceDate'].agg([('Recency','max')])
recency_df
```

```python
#merge 4 df into 1
customer_df=pd.concat([monetary_df,frequency_df,recency_df],axis=1)
customer_df["Recency"]=customer_df["Recency"].dt.days
customer_df
```

```python
#Add segent bins
customer_df['r_quartile'] = pd.qcut(customer_df['Recency'], 4, labels=list('4321'))
customer_df['f_quartile'] = pd.qcut(customer_df['Frequency'], 4,labels=list('321'),duplicates='drop')
customer_df['m_quartile'] = pd.qcut(customer_df['Monetary'], 4, labels=list('4321'),duplicates='drop')
customer_df[['r_quartile','f_quartile','m_quartile']]=customer_df[['r_quartile','f_quartile','m_quartile']].astype('int')
customer_df['RFM_Score'] = customer_df['r_quartile'] + customer_df['f_quartile'] + customer_df['m_quartile']
customer_df
```

```python
quantile = customer_df.quantile(q=[0.5,0.8])
quantile
```

```python
#Classify function
def segment(x):
    if x < quantile['RFM_Score'][0.5]: 
        return 'low-value'
    elif x > quantile['RFM_Score'][0.8]: #high_values class < 20%
        return 'high-value'
    else: return 'mid-value' #mid_class > 30%
```

```python
customer_df['Rank'] = customer_df['RFM_Score'].apply(segment)
customer_df
```

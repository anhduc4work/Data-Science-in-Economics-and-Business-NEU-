```python
c = dete_survey_updated.copy()
dete_resignations = c[c['separationtype'].str.contains('Resignation',
                                                       regex=False)]
# dete_resignations
```

```python
d = tafe_survey.copy()
tafe_resignations = d[d['Reason for ceasing employment'] == 'Resignation']
# tafe_resignations
```

```python
dete_resignations['cease_date'].value_counts()
```

```python
y = dete_resignations['cease_date'].str.extractall(r'(?P<Year>\d{4})')
dete_resignations['cease_date'] = y.reset_index(level=[1]).drop(
    'match', axis=1)['Year'].astype('float64') #turn multiindex to single index, synch with df indices
dete_resignations['cease_date']
```

```python
dete_resignations['cease_date'].value_counts()
```

```python
dete_resignations['dete_start_date'] = dete_resignations[
    'dete_start_date'].astype('float')
dete_resignations['dete_start_date'].value_counts()
```

```python
tafe_resignations['CESSATION YEAR'].value_counts()
```

```python
dete_resignations['institute_service'] = dete_resignations[
    'cease_date'] - dete_resignations['dete_start_date']
dete_resignations['institute_service']
```

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts(
    dropna=False)
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts(
    dropna=False)
```

```python
def update_vals(val):
    if val == 'Job Dissatisfaction' or val == 'Contributing Factors. Dissatisfaction ':
        return True
    elif val == '-':
        return False
    else:
        return np.nan


tafe_resignations[[
    'Contributing Factors. Dissatisfaction',
    'Contributing Factors. Job Dissatisfaction'
]] = tafe_resignations[[
    'Contributing Factors. Dissatisfaction',
    'Contributing Factors. Job Dissatisfaction'
]].applymap(update_vals)
```

```python
tafe_resignations['Dissatisfied'] = tafe_resignations[[
    'Contributing Factors. Dissatisfaction',
    'Contributing Factors. Job Dissatisfaction'
]].any(axis=1)
tafe_resignations_up = tafe_resignations.copy()
tafe_resignations_up['Dissatisfied'].value_counts()
```

```python
dete_resignations['Dissatisfied'] = dete_resignations[[
    'dissatisfaction_with_the_department', 'job_dissatisfaction'
]].any(axis=1)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['Dissatisfied'].value_counts()
```

```python
def exp_segment(y):
    if y < 1 and y >= 0:
        return 'Less than 1 year'
    elif y < 2:
        return '1-2'
    elif y < 5:
        return '3-4'
    elif y < 7:
        return '5-6'
    elif y < 11:
        return '7-10'
    elif y < 21:
        return '11-20'
    elif y >= 21:
        return 'More than 20 years'
    else:
        return np.nan
```

```python
dete_resignations_up['service_cat'] = dete_resignations_up[
    'institute_service'].apply(exp_segment)
dete_resignations_up['service_cat']
```
